# scraping

```python
import requests
from bs4 import BeautifulSoup

df = pd.DataFrame()

for i in codigo:
    req = requests.get(f'https://statusinvest.com.br/acoes/{i}')
    if req.status_code == 200:
        content = req.content

    soup = BeautifulSoup(content, 'html.parser')
    table = soup.find_all('div', {'class':'info'})
    x = soup.find_all('div', {'class':'w-50 w-sm-33 w-md-25 w-lg-16_6 mb-2 mt-2 item'})
    y = soup.find_all('div', {'class':'w-50 w-sm-33 w-md-25 w-lg-50 mb-2 mt-2 item'})

    dic = {'codigo' : i ,table[27].h3.text : table[27].strong.text, table[29].h3.span.text : table[29].strong.text,
          x[0].h3.text : x[0].strong.text, 
          x[1].h3.text : x[1].strong.text, 
          x[2].h3.text : x[2].strong.text,
          x[3].h3.text : x[3].strong.text, 
          x[4].h3.text : x[4].strong.text, 
          x[5].h3.text : x[5].strong.text,
          x[6].h3.text : x[6].strong.text, 
          x[7].h3.text : x[7].strong.text, 
          x[8].h3.text : x[8].strong.text,
          x[9].h3.text : x[9].strong.text, 
          x[10].h3.text : x[10].strong.text, 
          x[11].h3.text : x[11].strong.text,
          x[12].h3.text : x[12].strong.text, 
          x[13].h3.text : x[13].strong.text, 
          x[14].h3.text : x[14].strong.text,
          x[15].h3.text : x[15].strong.text,
          x[16].h3.text : x[16].strong.text,
          x[17].h3.text : x[17].strong.text,
          x[18].h3.text : x[18].strong.text,
          x[19].h3.text : x[19].strong.text,
          y[0].div.h3.text : y[0].div.strong.text,
          y[1].div.h3.text : y[1].div.strong.text,
          y[2].div.h3.text : y[2].div.strong.text,
          y[3].div.h3.text : y[3].div.strong.text}

    df = df.append(dic, ignore_index=True)
```

```python
df.head()
```

```python
for i in df.columns[:-1]:  #tiro a coluna codigo do loop   
    df[i] = df[i].str.replace('-%', '0')
    df[i] = df[i].str.replace('.', '')
    df[i] = df[i].str.replace(',', '.')
    df[i] = df[i].str.replace('-', '0')
    df[i] = df[i].str.replace('%', '').astype(float)
        
```

```python
df.info()
```

# regressao linear

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

x = df.drop(['Valor de mercado', 'codigo', 'Nº total de papeis'], axis = 1)
y = df['Valor de mercado']

# Dividindo X e Y em dados de treino e de teste
X_treino, X_teste, Y_treino, Y_teste = train_test_split(x, y, test_size = 0.26, random_state = 5) #test_size de 30% ideal
```

```python
regr = LinearRegression()
# Treinando o modelo
regr.fit(X_treino, Y_treino)
```

```python
regr.score(X_teste, Y_teste)
```

```python
atual = []
previsto = []

for i in range(len(df)):
    reg = regr.predict([x.loc[i]])
    xd = reg / df['Nº total de papeis'][i]
    previsto.append(xd.tolist()[0]) # tirando o array
    atual.append(df['Valor de mercado'][i] / df['Nº total de papeis'][i])

    
df['preco_atual'] = atual
df['preco_previsto'] = previsto
df['potencial de ganho (%)'] = (df['preco_previsto'] - df['preco_atual']) / df['preco_atual'] * 100
```

```python
maiores = df.nlargest(10, 'potencial de ganho (%)')
menores = df.nsmallest(10, 'potencial de ganho (%)')

for i in range(len(maiores)):
    print('{} | codigo: {} / maiores retornos: {:.2f}%'.format(i+1, maiores['codigo'].iloc[i], maiores['potencial de ganho (%)'].iloc[i]))
print('-----------------------')    
for i in range(len(maiores)):
    print('{} | codigo: {} / menores retornos: {:.2f}%'.format(i +1, menores['codigo'].iloc[i], menores['potencial de ganho (%)'].iloc[i]))
         
```
